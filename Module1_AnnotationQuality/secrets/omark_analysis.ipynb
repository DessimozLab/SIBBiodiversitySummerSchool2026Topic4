# generate OMArk report
First of all, we will generate the standard OMArk report for all 10 species we analysed. This plot summarises most of the stats calculated by OMArk in a 3-way stacked barplot per species proteome.
- The top-most bar plot shows the total number of proteins in the proteome. Here you can immediately notice deviations from the general average.
- In the second bar plot, you can observe the completeness of the proteome. Taking the chosen reference ancestral gene pool and comparing it to this proteome, what is the proportion of the ancestral genes that is found (in single or multiple copy) or missing. A high proportion of missing genes could be indication of an incomplete genome annotation, or, biologically, a high gene loss that is typical of endosymbiosis and/or parasitism.
- In the third bar plot, you can observe the consistency of the proteome. Here you can observe the proportion of extant gene that is found in the reference ancestral gene pool (consistent), or that maps to other organisms (inconsistent, unknown, or contamination). Additionally, you can observe how many genes are present as fragments.

# Filter by contaminants
Here we will apply different filters to select the proteomes of high quality only
e.g. by contaminants

```python
# add column with contaminants and fragments
omark_df['Contaminant_Total'] = omark_df['Contaminant_Partially_Mapping'] + omark_df['Contaminant_Fragments'] + omark_df['Contaminant_Structurally_Consistent']
omark_df['Fragments_Total'] = omark_df['Consistent_Fragments'] + omark_df['Inconsistent_Fragments'] + omark_df['Contaminant_Fragments']
omark_df['Partial_Mapping_Total'] = omark_df['Consistent_Partially_Mapping'] + omark_df['Inconsistent_Partially_Mapping'] + omark_df['Contaminant_Partially_Mapping']
omark_df['Inconsistent_Total'] = omark_df['Inconsistent'] + omark_df['Inconsistent_Fragments'] + omark_df['Inconsistent_Partially_Mapping'] + omark_df['Inconsistent_Structurally_Consistent']
```

```python
#basic filters
def basic_filter(df):
    df = prot_num_filter(df,80000)
    df = missing_filter(df,80)
    df = contami_filter(df,0)
    df = fragments_filter(df,30)
    df = inconsistent_filter(df,10)
    return df

def prot_num_filter(df,thr=100000):
    return df.loc[df['Protein_number']<thr]

def missing_filter(df,thr=90):
    return df.loc[df['Missing']<thr]

def contami_filter(df,thr=0):
    return df.loc[df['Contaminant_Total']<=thr]

def fragments_filter(df,thr=40):
    return df.loc[df['Fragments_Total']<thr]

def inconsistent_filter(df,thr=40):
    return df.loc[df['Inconsistent_Total']<thr]
```

```python
filtered_omark_df = basic_filter(omark_df)
filtered_omark_df
```
