# But first... what Is Genome Annotation?

Genome annotation is the process of identifying and labeling biologically meaningful elements in a genome sequence.

This usually includes:

## Structural Annotation

Detection of genomic features such as:

- protein-coding genes
- exons/introns boundaries
- untranslated regions (UTRs)
- non-coding RNAs
- pseudogenes

## Functional Annotation

Assigning biological meaning to predicted genes or proteins, such as:

- Gene Ontology (GO) terms, subdivided into:
    - Molecular Function (MF)
    - Biological Process (BP)
    - Cellular Component, or localization (CC)
- protein domains
- pathway membership, or enzymatic classification
- orthologous relationships

## Output Products

Genome annotation commonly produces:

- gene models (GFF/GTF files)
- transcript sequences
- protein sequences (proteomes)

In this notebook, we focus mainly on **predicted proteomes**, since many downstream comparative analyses rely on protein sequences.

# Why Annotation Quality Is Important

Even with a good genome assembly, poor annotation can strongly affect downstream analyses.

Common annotation problems include:

- missing genes
- fragmented, truncated, or duplicated proteins
- fused neighboring genes
- incorrect exon boundaries
- contamination or foreign sequences

These errors propagate into many analyses.

## Examples of Impact

### Orthology Inference

Orthology methods assume that gene models correctly represent real proteins.  
Missing or fragmented genes may be interpreted as gene loss.

### Functional Annotation

Incorrect protein sequences can lead to:

- missing domains
- wrong GO terms
- false pathway assignments

### Species Tree Inference

Phylogenomics often uses single-copy orthologs.  
Annotation errors can reduce gene recovery or introduce paralogs, biasing trees.

### Gene Family Evolution

False duplications or missing genes distort expansion/contraction analyses.

# Existing Tools for Annotation Quality Assessment

Several tools evaluate annotation quality from different perspectives.

## Completeness-Based Tools

### BUSCO

Measures presence of expected universal single-copy orthologs.

Useful for:

- completeness estimation
- comparing annotations
- selecting best assemblies/annotations

## Orthology-Based Tools

### OMArk

Uses Hierarchical Orthologous Group (HOG) databases to assess whether proteins are:

- expected
- missing
- duplicated
- fragmented
- taxonomically inconsistent (putative contaminations)

## Additional Common Approaches (not included directly)

- Other completeness tools (EukCC, DOGMA, CheckM)
- UniProt Complete Proteome Detector
- Read mapping support (RNA-seq evidence)
- Protein homology support
- Manual curation
- Gene structure statistics
- Contamination screening

No single metric captures all annotation quality dimensions, so combining tools is recommended.

# Conceptual Principle of BUSCO

BUSCO uses curated sets of genes that are expected to exist as **single-copy orthologs** in nearly all members of a lineage.

For birds or vertebrates, many genes are highly conserved and usually present once per genome.

BUSCO searches the proteome and classifies each expected gene as:

- **Complete (Single-copy)** → found once and intact
- **Complete (Duplicated)** → found more than once
- **Fragmented** → partial match detected
- **Missing** → not detected

## Interpretation

High-quality annotations generally show:

- high Complete %
- low Fragmented %
- low Missing %

Large duplicated fractions may indicate redundant annotations or recent duplications.

# Conceptual Principle of OMArk

OMArk evaluates proteomes using orthology knowledge derived from databases such as the OMA database.

Instead of only checking conserved single-copy genes, OMArk examines whether predicted proteins fit expected evolutionary patterns.

## OMArk Can Detect

- conserved genes that should be present but are absent
- duplicated genes
- partial proteins
- lineage-inconsistent proteins
- possible contamination
- annotation artifacts

## Why This Is Valuable

Two proteomes may have similar BUSCO completeness scores but differ strongly in biological consistency.  
OMArk helps diagnose these subtler issues.

## Complementarity with BUSCO

- **BUSCO** answers: *How complete is the annotation?*
- **OMArk** answers: *How evolutionarily consistent and reliable is the proteome?*

Using both gives a broader view of annotation quality.
